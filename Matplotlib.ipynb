```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
max_t = 100
t = np.arange(0, max_t, 5)
a = 52
b = 2

def f(t):
    return np.exp(-(t/a)**b)

plt.plot(t, f(t))
plt.title(r'De-rating Factor')
plt.xlabel('Age (Years)')
plt.ylabel('Age De-rating factor')
plt.axis([0, max_t, 0, 1])
plt.grid(True)
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# Bar chart

data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

# single plot:
plt.bar(names, values)
plt.show()
```

```python
# Hystograms ------------------

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# 3 plot in a figure:
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
plt.show()
```

```python
# Working with multiple figures and axes:

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Plotting with Pandas

```python
import pandas as pd
```

```python
# Series plot

ser = pd.Series(['c','c','b','a','b','c','b','a','a','b','c','c'])

ser_c = ser.value_counts()
s=ser_c.sort_index()
s.plot(kind='bar')
plt.show()
```

```python
ser.value_counts().plot(kind='bar', grid=True) # also works
ind = np.arange(3)
#plt.xticks(ind,('a','b','c'))
plt.show()
```

```python
ser_n = pd.Series([2,2,1,4,3,4,4,1,2])
plt.hist(ser_n)
plt.show()
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0,100,10))
s.plot()
plt.show()
```

```python
# DataFrame plot

df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                  columns=['A','B','C','D'],
                  index=np.arange(0,100,10))
df.plot(figsize = (8,5))
plt.show()
```
